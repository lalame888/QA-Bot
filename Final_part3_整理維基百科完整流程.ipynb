```
from pyspark import SparkContext
from pyspark import SparkConf
sc =SparkContext()
```

```
Wiki=sc.textFile("/content/drive/My Drive/Wiki_inverse_index")
```

```
Wiki.take(1)
```

```
A=Wiki.map(lambda line : line.split(" "))
```

```
A.take(2)
```

```
def filt(x):
  result=[]
  for t in x:
    if t != " " and t!="　" and t!="":
      result.append(t)

  
  return result
```

```
A11=A.map(filt)
```

```
A11.take(1)
```

```
def bulidList(x):
  
  listA=[]
  for t in x[1:]:
    listA.append(int(t))

  new_list=list(set(listA))   #把重複的刪掉
  new_list.sort(reverse=False)

  txt=""
  txt+=x[0]+' '


  for n in new_list:
    txt+=str(n)+" "
  #ans=(x[0],txt)
  
  return txt
```

```
A1=A.map(bulidList)
```

```
A1.take(1)
```

```
A2=A1.collect()
```

```
txt=""
for x in A2:
  
  txt+=str(x)
  txt+='\n'

import re
fp = open("/content/drive/My Drive/Wiki_inverse_index", "w")
fp.write(txt)

fp.close()
```
