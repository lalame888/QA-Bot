```
import os, sys
os.environ['SPARK_HOME'] = "/usr/local/spark"
os.environ['PYSPARK_PYTHON'] = "/usr/local/bin/python"
sys.path.append("/usr/local/spark/python/")
sys.path.append("/usr/local/spark/python/lib/pyspark.zip")
sys.path.append("/usr/local/spark/python/lib/py4j-0.10.4-src.zip")
```

```
from pyspark import SparkContext
from pyspark import SparkConf
sc =SparkContext()
```

```
AllWiki=sc.textFile("/content/drive/My Drive/AllWiki")
```

```
AllWiki.take(1)
```

```
A=AllWiki.map(lambda line : line.split(" "))
```

```
A.take(1)
```

```
def AllFilt(x):
  result=[]
  result.append(x[0])
  result.append(x[1])
  result.append(x[2])
  for t in x[3:]:

    if t.startswith("n")  or t.startswith("a") or t.startswith("d") or t.startswith("i") or t.startswith("m") or t.startswith("q") or t.startswith("v") or t.startswith("v"):
      try:
        result.append(t.split("/")[1])
      except:
        pass
  return result
```

```
All=A.map(AllFilt)
```

```
All.take(1)
```

```
def filt(x):
  result=[]
  for t in x:
    if t != "，" and t!="。" and t!="的" and t!="了" and t!="是" and t!="他" and t!="她" and t!="它" and t!="（" and t!="）" and t!="、" and t!="」"  and t!="「"  :
      result.append(t)
  return result

  
```

```
All2=All.map(filt)
```

```
All2.take(1)
```

```
def com(x):
  text=""
  for t in x:
    text+=" "
    text+=t
  return text
```

```
All3=All2.map(com)
```

```
All3.take(10)
```

```
Wiki=All3.collect()
```

```
txt=""
for x in Wiki:
  txt+=x
  txt+='\n'
```

把整理好的WIKI存回去硬碟

```
import re
fp = open("/content/drive/My Drive/Wiki", "w")
fp.write(txt)

fp.close()
```
